```python
import json
import pandas as pd
```

```python
df = pd.read_csv("CykloRank města - města nad 10000, openstreet ID.csv")
```

```python
city_mappings = {
    'europe/czech-republic-latest.osm.pbf': [],
}

for row in df.iterrows():
    print(row[1]['jméno'], row[1]['openstreet ID'], row[1]['GPS'])
    GPS = tuple([float(x) for x in row[1]['GPS'].split(", ")])
    jmeno = row[1]['jméno'].replace(" ", "_")
    city_mappings['europe/czech-republic-latest.osm.pbf'].append(
    {jmeno: {'osm_id': row[1]['openstreet ID'], 'centre': GPS}},
    )
```

```python
city_mappings
```

```python
json.dump(city_mappings, open("city_mappings_cz.py", "w"), indent=4)
```
